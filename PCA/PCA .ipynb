##### Perform Principal component analysis and perform clustering using first 3 principal component scores (both heirarchial and k mean clustering(scree plot or elbow curve) and obtain optimum number of clusters and check whether we have obtained same number of clusters with the original data (class column we have ignored at the begining who shows it has 3 clusters)df

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import warnings
warnings.filterwarnings('ignore')
```

```python
wine_df = pd.read_csv('C:/Users/soumy/documents/PCA/wine.csv')
wine_df
```

```python
wine_df['Type'].value_counts()
```

```python
#checking null values
wine_df.isna().sum()
```

```python
wine1=wine_df.iloc[:,1:]
wine1
```

```python
wine1.shape
```

```python
wine1.info()
```

```python
wine_ary = wine1.values
wine_ary
```

```python
wine_norm = scale(wine_ary)
wine_norm
```

```python
pca=PCA(n_components=13)
wine_pca=pca.fit_transform(wine_norm)
wine_pca
```

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
plt.plot(var1,color='blue')
```

```python
final_df = pd.concat([wine_df['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
fig=plt.figure(figsize=(15,12))
sns.scatterplot(data=final_df, color='purple')
plt.show()

```

### Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,8))
dendrogram = sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
hclusters = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y = pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
wine2 = wine_df.copy()
wine2['clustersid']=hclusters.labels_
wine2
```

### K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range (1,6):
    kmeans = KMeans(n_clusters=i,random_state=60)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python

clusters = KMeans(3,random_state=50).fit(wine_norm)
clusters
```

```python
clusters.labels_
```

```python
wine3 = wine_df.copy()
wine3['clustersid']=clusters.labels_
wine3
```
